```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../Datasets/groceries.csv')
data.drop('Item(s)', axis=1, inplace=True)
data.head()
```

```python
data.shape
```

```python
market = []
for i in range(0, data.shape[0]):
    cus = []
    for j in data.columns:
        if type(data[j][i]) == str:
            cus.append(data[j][i])
    market.append(cus)
```

```python
data_list = []
for i in market:
    for j in i:
        data_list.append(j)
```

```python
import collections
```

```python
no_items = collections.Counter(data_list)
```

```python
d = {'Item Name': no_items.keys(), 'Values': no_items.values()}
```

```python
pd.DataFrame(d).sort_values(by=['Values'], ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
df = pd.DataFrame(tr.transform(market), columns=tr.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
support_items = apriori(df, min_support=0.05, use_colnames=True, max_len=3).sort_values(by=['support'], ascending=False)
```
