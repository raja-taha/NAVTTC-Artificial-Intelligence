```python
import pandas as pd
import numpy as np
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
# Fix np.NINF issue by replacing it with -np.inf
if not hasattr(np, 'NINF'):
    np.NINF = -np.inf
    
data = pd.read_csv('../Datasets/diabetes.csv')
data.head()
```

```python
# x = data.drop(['Pregnancies', 'Insulin', 'DiabetesPedigreeFunction' ,'Outcome'], axis=1)
x = data.drop('Outcome', axis=1)
y = data['Outcome']
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
fs = SequentialFeatureSelector(lr, k_features=8, forward=True)
```

```python
fs.fit(x, y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
