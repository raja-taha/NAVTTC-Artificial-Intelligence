```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/loan.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.distplot(data['CoapplicantIncome'])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(data[['CoapplicantIncome']])
```

```python
data['CoapplicantIncome_ms'] = ms.transform(data[['CoapplicantIncome']])
```

```python
data.head()
```

```python
plt.subplot(1, 2, 1)
plt.title("Before Normalization")
sns.distplot(data['CoapplicantIncome'])
plt.subplot(1, 2, 2)
plt.title("After Normalization")
sns.distplot(data['CoapplicantIncome_ms'])
plt.show()
```
