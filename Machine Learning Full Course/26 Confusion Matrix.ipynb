```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/Social_Network_Ads.csv')
data.drop(['User ID', 'Gender'], axis=1, inplace=True)
data.head()
```

```python
x = data.iloc[:, :-1]
y = data['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
cf = confusion_matrix(y_test, lr.predict(x_test))

sns.heatmap(cf, annot=True)
plt.show()
```

```python
print("Accuracy Score: ", accuracy_score(y_test, lr.predict(x_test))*100)
print("Precision Score: ", precision_score(y_test, lr.predict(x_test))*100)
print("Recall Score: ", recall_score(y_test, lr.predict(x_test))*100)
print("F1 Score: ", f1_score(y_test, lr.predict(x_test))*100)
```
