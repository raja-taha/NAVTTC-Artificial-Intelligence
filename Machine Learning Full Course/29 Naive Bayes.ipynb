```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv('../Datasets/Social_Network_Ads.csv')
data.drop(['User ID', 'Gender'], axis=1, inplace=True)
data.head()
```

```python
sns.scatterplot(x='Age', y='EstimatedSalary', data=data, hue='Purchased')
plt.show()
```

```python
sns.kdeplot(data=data['Age'], fill=True)
plt.show()
```

```python
sns.kdeplot(data=data['EstimatedSalary'], fill=True)
plt.show()
```

```python
sns.kdeplot(data=data['Purchased'], fill=True)
plt.show()
```

```python
x = data.iloc[:, :-1]
y = data['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
gnb.score(x_test, y_test)*100, gnb.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=gnb)
plt.show()
```

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
mnb.score(x_test, y_test)*100, mnb.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=mnb)
plt.show()
```

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
bnb.score(x_test, y_test)*100, bnb.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=bnb)
plt.show()
```
