```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Datasets/loan.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data['ApplicantIncome'].fillna(data['ApplicantIncome'].mean(), inplace=True)
```

```python
sns.distplot(data['ApplicantIncome'])
plt.show()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(data[['ApplicantIncome']])
```

```python
data['ApplicantIncome_ss'] = pd.DataFrame(ss.transform(data[['ApplicantIncome']]), columns=['x'])
```

```python
data.head()
```

```python
data.describe()
```

```python
plt.subplot(1, 2, 1)
plt.title("Before Standardization")
sns.distplot(data['ApplicantIncome'])
plt.subplot(1, 2, 2)
plt.title("After Standardization")
sns.distplot(data['ApplicantIncome_ss'])
plt.show()
```
